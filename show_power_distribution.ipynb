```python
import torch
from power_spherical import HypersphericalUniform, PowerSpherical

import seaborn as sns
import matplotlib.pyplot as plt
import math

dist = torch.distributions.normal.Normal(torch.tensor([0.0]), torch.tensor([1.0]))

sns.kdeplot(dist.sample((1000000,)))
```

```python
dist = HypersphericalUniform(dim=1)

sns.kdeplot(dist.sample((100000,)))
```

```python
x = dist.sample(torch.Size([100000]))

plt.hist((lambda v: torch.where(v > 0, v, 2 * math.pi + v))(torch.atan2(x[:,1], x[:,0])),
         density=True, bins=50, alpha=.3, label='samples')

plt.legend()
plt.show()
```

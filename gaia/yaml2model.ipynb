```python
from spherinator.models import yaml2model
import torch

model = yaml2model("gaia_ae_8_trunc_normal.yaml")
model(torch.randn(1, 1, 343))
model
```

```python
# Print parameter summary
print("Model Parameter Summary:")
print("-" * 80)
print(f"{'Module':<40} {'Shape':<20} {'Requires Grad':<15} {'Params':<10}")
print("-" * 80)

total_params = 0
trainable_params = 0

for name, param in model.named_parameters():
    param_count = param.numel()
    total_params += param_count
    if param.requires_grad:
        trainable_params += param_count

    print(
        f"{name:<40} {str(list(param.shape)):<20} {str(param.requires_grad):<15} {param_count:<10,}"
    )

print("-" * 80)
print(f"Total parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")
print(f"Non-trainable parameters: {total_params - trainable_params:,}")
```

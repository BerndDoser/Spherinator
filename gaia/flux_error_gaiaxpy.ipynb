```python
import matplotlib.pyplot as plt
from gaiaxpy import calibrate
import numpy as np

# id = 38655544960  # first entry
# id = 2947050466531872640   # Paper Fig. 22
# id = 2947047202356748672
id = 194912732657384320

# sampling = np.arange(336, 1021, 2)
sampling = np.arange(336, 1021, 10)

data, _ = calibrate([id], sampling=sampling, save_file=False)

flux = data["flux"][0]
flux_error = data["flux_error"][0]
```

```python
plt.errorbar(range(len(flux)), flux, yerr=flux_error, ecolor="red")
```

```python
data_min = flux.min()
data_max = flux.max()

flux = (flux - data_min) / (data_max - data_min)
flux_error = flux_error / (data_max - data_min)
```

```python
plt.errorbar(range(len(flux)), flux, yerr=flux_error, ecolor="red")
```

```python
plt.plot(range(len(flux_error)), flux_error)
plt.xlabel("Index")
plt.ylabel("Flux Error")
plt.title("Flux Error Plot")
plt.show()
```

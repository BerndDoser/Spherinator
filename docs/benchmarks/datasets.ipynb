```python
import pandas as pd

import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("datasets.csv")

# Convert bytes to MB for better readability
df["Size [MB]"] = df["Size [bytes]"] / (1024 * 1024)

# Sort by size in descending order
df = df.sort_values("Size [MB]", ascending=False)

# Create a bar plot instead of scatter plot
plt.figure(figsize=(10, 6))
plt.bar(df["Format"], df["Size [MB]"])
plt.xlabel("Format")
plt.ylabel("Size (MB)")

# Add size labels on top of each bar
for i, v in enumerate(df["Size [MB]"]):
    plt.text(i, v + max(df["Size [MB]"]) * 0.01, f"{v:.1f}", ha="center", va="bottom")
plt.title("Size vs Format")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
